Consolidando os dfs

```python
def consolidar_dados(df_principal, df_secundario):
    # Realiza a consolidação usando a função merge do pandas
    df_consolidado = pd.merge(df_principal, df_secundario, on='CNPJ_CIA', how='inner')  # 'CNPJ' é a chave de junção

    return df_consolidado
```

```python
# Realiza o left join entre df_ativo_total e df_ativo_total_trim usando CNPJ_CIA como chave
ativo_total_consolidado = pd.merge(df_ativo_total, df_ativo_total_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_receita_liquida e df_receita_liquida_trim usando CNPJ_CIA como chave
receita_liquida_consolidado = pd.merge(df_receita_liquida, df_receita_liquida_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_patrimonio_liquido e df_patrimonio_liquido_trim usando CNPJ_CIA como chave
patrimonio_liquido_consolidado = pd.merge(df_patrimonio_liquido, df_patrimonio_liquido_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_divida_liquida e df_divida_liquida_trim usando CNPJ_CIA como chave
divida_liquida_consolidado = pd.merge(df_divida_liquida, df_divida_liquida_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_free_cash_flow e df_free_cash_flow_trim usando CNPJ_CIA como chave
free_cash_flow_consolidado = pd.merge(df_free_cash_flow, df_free_cash_flow_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_cx_liq_ativ_financ e df_cx_liq_ativ_financ_trim usando CNPJ_CIA como chave
cx_liq_ativ_financ_consolidado = pd.merge(df_cx_liq_ativ_financ, df_cx_liq_ativ_financ_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_receita_liquida e df_receita_liquida_trim usando CNPJ_CIA como chave
receita_liquida_consolidado_2 = pd.merge(df_receita_liquida, df_receita_liquida_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_EBTIDA e df_EBTIDA_trim usando CNPJ_CIA como chave
EBTIDA_consolidado = pd.merge(df_EBTIDA, df_EBTIDA_trim, on='CNPJ_CIA', how='left')

# Realiza o left join entre df_lucro_liquido e df_lucro_liquido_trim usando CNPJ_CIA como chave
lucro_liquido_consolidado = pd.merge(df_lucro_liquido, df_lucro_liquido_trim, on='CNPJ_CIA', how='left')


# Removendo a coluna 'DENOM_CIA_y' de cada DataFrame
receita_liquida_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
patrimonio_liquido_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
divida_liquida_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
free_cash_flow_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
cx_liq_ativ_financ_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
receita_liquida_consolidado_2.drop('DENOM_CIA_y', axis=1, inplace=True)
EBTIDA_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
lucro_liquido_consolidado.drop('DENOM_CIA_y', axis=1, inplace=True)
```

```python
# Dicionário para mapear cada DataFrame e sua observação correspondente
dfs = {
    'Ativo Total': ativo_total_consolidado,
    'Patrimonio Liquido': patrimonio_liquido_consolidado,
    'Divida líquida': divida_liquida_consolidado,
    'Fluxo caixa livre - free cash flow': free_cash_flow_consolidado,
    'Cx liq ativid de financeiamento': cx_liq_ativ_financ_consolidado,
    'Receita liquida': receita_liquida_consolidado,
    'EBTIDA': EBTIDA_consolidado,
    'Lucro liquido': lucro_liquido_consolidado
}

# Diretório para salvar os arquivos
diretorio_base = caminho_salvar,

dfs = {
    'Ativo Total': ativo_total_consolidado,
    'Patrimonio Liquido': patrimonio_liquido_consolidado,
    'Divida líquida': divida_liquida_consolidado,
    'Fluxo caixa livre - free cash flow': free_cash_flow_consolidado,
    'Cx liq ativid de financeiamento': cx_liq_ativ_financ_consolidado,
    'Receita liquida': receita_liquida_consolidado,
    'EBTIDA': EBTIDA_consolidado,
    'Lucro liquido': lucro_liquido_consolidado
}

# Diretório para salvar os arquivos
diretorio_base = caminho_salvar,

# Iterar sobre cada linha do DataFrame df_empresas
for index, row in df_empresas.iterrows():
    # Obtendo o CNPJ e o nome da empresa da linha atual
    cnpj_empresa = row['CNPJ']  # Coloque o nome da coluna correta do CNPJ
    nome_empresa = row['Empresa']  # Coloque o nome da coluna correta do nome da empresa

    # Lista para armazenar os DataFrames de cada observação
    dfs_empresa = []

    # Iterar sobre as chaves do dicionário dfs
    for observacao, df_consolidado in dfs.items():
        # Filtrar o DataFrame pelo CNPJ da empresa
        df_temp = df_consolidado[df_consolidado['CNPJ_CIA'] == cnpj_empresa].copy()

        # Selecionar apenas colunas numéricas a partir da terceira coluna
        numeric_cols = df_temp.columns[2:]
        df_temp[numeric_cols] = df_temp[numeric_cols].apply(pd.to_numeric, errors='coerce')

        # Dividir os dados por 1000
        df_temp[numeric_cols] /= 1000  # Dividindo as colunas numéricas por 1000

        # Excluir as duas primeiras colunas do DataFrame
        df_temp.drop(df_temp.columns[[0, 1]], axis=1, inplace=True)

        # Remover a coluna 'DENOM_CIA_y' se ela existir
        if 'DENOM_CIA_y' in df_temp.columns:
            df_temp.drop('DENOM_CIA_y', axis=1, inplace=True)

        # Adicionar a coluna com o nome da empresa
        df_temp['Nome Empresa'] = nome_empresa

        # Adicionar a observação ao DataFrame
        df_temp['Observação'] = observacao

        # Reorganizar as colunas colocando 'Observação' na primeira posição
        cols = df_temp.columns.tolist()
        cols = ['Observação'] + [col for col in cols if col != 'Observação']
        df_temp = df_temp[cols]

        # Renomear a coluna 'Preço atual vs preço intrínseco' para 'Observação'
        df_temp.rename(columns={'Observação': 'Preço atual vs preço intrínseco'}, inplace=True)

        # Adicionar o DataFrame ao lista
        dfs_empresa.append(df_temp)

    # Concatenar os DataFrames de cada observação para a empresa atual
    df_empresa_observacoes = pd.concat(dfs_empresa, axis=0, ignore_index=True)

    # Salvar o DataFrame no diretório específico com o nome do CNPJ
    nome_arquivo_excel = f"{diretorio_base}/{nome_empresa}.xlsx"
    df_empresa_observacoes.to_excel(nome_arquivo_excel, index=False)
```
